## **Matplotlib Beginner friendly Notebook :)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

Simple Plotting

```python
plt.title('Gas Prices over time (in USD)')          # The title :) pretty starightforward right
plt.plot(gas.Year, gas.USA, label = 'USA')

plt.plot(gas.Year,gas.Canada, label = 'Canada')       #plot another country with respect to USA You can also use gas['Year] instead of gas.Year :)
plt.plot(gas.Year, gas['South Korea'], label ='South Korea')

plt.legend()                                        # to show the legend(You need to add label inside the plot() in order to show the legend)
plt.show()
```

Some other things that you can do

```python
plt.title('Gas Prices over time (in USD)')
plt.plot(gas.Year, gas.USA, 'b.-' , label = 'USA' )
plt.plot(gas.Year,gas.Canada, 'r.-')                                 #  b = indicates blue lines
plt.plot(gas.Year, gas['South Korea'], 'g.-')                        #  . = indicates dotted
plt.plot(gas.Year,gas.Canada,'r.-', label = 'Canada' )               #  - = indicates line joining points
plt.plot(gas.Year, gas['South Korea'], 'g.-' ,label ='South Korea')
# You can also do these things

print(gas.Year[::3])
plt.xticks(gas.Year[::3])

plt.xlabel('Years')             # Labelling x, y
plt.ylabel('US Dollars')

plt.legend()
plt.show()       # The Year difference will be three years now.
```

Print all the countries in a smart way

```python
plt.figure(figsize = (8,5))
plt.title('Gas Prices over time (in USD)' , fontdict = {'fontweight' :'bold'})   # Your Title is Bolder now

for country in gas:
  if country !='Year':
    plt.plot(gas.Year, gas[country], marker = '.')
plt.legend(gas)
plt.show
```

## **Load Fifa Dataset**

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

**Histograms**

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins = bins, color = '#8a9ff2' )   # You can modify things in various ways
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')
plt.show()
```

**Pie Charts**

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['left','right']
colors = ['#abcdef','#aabbcc']

plt.pie([left,right],labels = labels,colors = colors , autopct = '%.2f %%') #%% is for showing the symbol on the pie charts (syntax)
plt.title('Foot Preference of FIFA Players')
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight.head(5)
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
medium_heavy
```

```python
weights = [light,medium,light_medium,medium_heavy,heavy]
plt.style.use('ggplot')  # using a default style from module
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,0.2,0,0,0.4)
plt.title('Weight Distribution of FIFA Players')
plt.pie(weights,labels = labels, autopct = '%.2f %%',explode = explode)
plt.show()
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']
plt.boxplot([barcelona,madrid, revs], labels = labels)
plt.title('Comparison of FC Barcelona and Real Madrid')
plt.show()
```
