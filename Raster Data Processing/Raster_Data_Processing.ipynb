```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
import numpy as np

# Read the raster data
file_path = '/content/drive/MyDrive/GirlsScriptOpenSource/Mlrepo/Raster Data Processing/synthetic_raster.tif'
with rasterio.open(file_path) as src:
    raster = src.read(1)  # Read the first band

# Display the raster data using imshow to get a mappable object
plt.figure(figsize=(10, 10))
plt.imshow(raster, cmap='viridis')  # Use imshow instead of show
plt.title('Raster Data')
plt.colorbar(label='Pixel Value') # Now the colorbar will work
plt.show()
```

```python
# Simple processing example: Normalization
raster_normalized = (raster - raster.min()) / (raster.max() - raster.min())

# Display the normalized raster data
plt.figure(figsize=(10, 10))
plt.imshow(raster_normalized, cmap='viridis')
plt.title('Normalized Raster Data')
plt.colorbar(label='Normalized Pixel Value')
plt.show()
```

```python
# Example of a more complex operation: Thresholding
threshold = 0.5
raster_thresholded = np.where(raster_normalized > threshold, 1, 0)

# Display the thresholded raster data
plt.figure(figsize=(10, 10))
plt.imshow(raster_thresholded, cmap='gray')
plt.title('Thresholded Raster Data')
plt.colorbar(label='Binary Value')
plt.show()
```
