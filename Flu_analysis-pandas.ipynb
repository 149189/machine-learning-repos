<p><a id="labels"></a></p>
<h2>Labels</h2>
<hr />
<p>For this competition, there are two target variables:</p>
<ul>
<li><code>h1n1_vaccine</code> - Whether respondent received H1N1 flu vaccine.</li>
<li><code>seasonal_vaccine</code> - Whether respondent received seasonal flu vaccine.</li>
</ul>
<p>Both are binary variables: <code>0</code> = No; <code>1</code> = Yes. Some respondents didn't get either vaccine, others got only one, and some got both. This is formulated as a multilabel (and <em>not</em> multiclass) problem.</p>
<p><a id="features_list"></a></p>

<h2>Performance metric</h2>
<hr />
<p>Performance will be evaluated according to the area under the receiver operating characteristic curve (ROC AUC) for each of the two target variables. The mean of these two scores will be the overall score. A higher value indicates stronger performance.</p>
<p>In Python, you can calculate this using <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_auc_score.html"><code>sklearn.metrics.roc_auc_score</code></a> for this multilabel setup with the default <code>average="macro"</code> parameter.</p>
<p><a id="sub_values"></a></p>
<h2>Submission format</h2>
<hr />
<p>The format for the submission file is three columns: <code>respondent_id</code>, <code>h1n1_vaccine</code>, and <code>seasonal_vaccine</code>. The predictions for the two target variables should be <strong>float probabilities</strong> that range between <code>0.0</code> and <code>1.0</code>. Because the competition uses ROC AUC as its evaluation metric, the values you submit must be the probabilities that a person received each vaccine, <em>not</em> binary labels.</p>
<p>As this is a multilabel problem, the probabilities for each row do <em>not</em> need to sum to one.</p>
