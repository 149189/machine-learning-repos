```python
#importing the neccesory library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon
```

```python
#Read the CSV file
data = pd.read_csv('/content/drive/MyDrive/GirlsScriptOpenSource/ml|RepoProjects/Wilcoxon Signed-Rank Test/paired_data.csv')
```

```python

#Perform Wilcoxon Signed-Rank Test
statistic, p_value = wilcoxon(data['before'], data['after'])
```

```python
# Display results
print(f"Wilcoxon Signed-Rank Test:")
```

```python
print(f"Statistic: {statistic}")
```

```python
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value >= alpha:
    print(f"Since p-value ({p_value:.4f}) >= alpha ({alpha}), we do not reject the null hypothesis.")
else:
    print(f"Since p-value ({p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))

plt.hist(data['before'], alpha=0.5, label='Before')
plt.hist(data['after'], alpha=0.5, label='After')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Before and After Measurements')
plt.legend()
```

```python
# Highlight significant difference if p-value < 0.05
alpha = 0.05
if p_value < alpha:
    plt.text(0.95, 0.95, "Significant Difference", ha='right', va='top', transform=plt.gca().transAxes, bbox=dict(facecolor='red', alpha=0.5))

plt.show()
```
